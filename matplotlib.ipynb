```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('title')
```

```python
'string'
```

```python
plt.subplot(1, 2, 1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('title')
```

```python
fig = plt.figure()

axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis2 = fig.add_axes([0.5, 0.15, 0.4, 0.3])

axis1.plot(x,y)
axis2.plot(y,x)

axis1.set_title('outer')
axis2.set_title('inner')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].set_title('First One')
axes[1].set_title('Second One')

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X-square')
ax.plot(x,x**3,label='Y-square')

ax.legend(loc=10)
```

```python
fig.savefig('my_picture.png')
```

```python
fig = plt.figure()

axe = fig.add_axes([0,0,1,1])

axe.plot(x,y, color='#FF8C00', linewidth=3, linestyle='-', marker='o', markersize=3, markerfacecolor='black', markeredgecolor='red') 

axe.set_xlim([0,1])
axe.set_ylim([0,2])
```
